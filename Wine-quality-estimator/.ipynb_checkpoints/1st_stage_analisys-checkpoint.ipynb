## Notes

#### **Features of the wine showing**:

* **No relation to the quality**:
    * fixed acidity, residiual sugar
<br/><br/>
* **To be examined further**:
    * density
    * pH
<br/><br/>
* **Uncertain relation to the quality**:
    * citric acid - somewhat of an upward trend vs quality (variance is a bit high though) - SHOW
    * free sulfur dioxide - there could be a trend, but the variance is too high
    * density - there could be small variations - checked (there could be a trend but variance too high)
    * pH - big var but there could be something - SCK
<br/><br/>
* **More certain relation to the quality**:
    * volatile acidity - clear downward trend vs quality - SHOW
    * chlorides - the variation in the chlorides gets linearly lower with the increase of wine quality - SHOW
    * sulfates - good upward trend, though var swh high - SHOW
    * alchohol - slight but defined upward trend - SHOW

```python
# make pie chard with High, Mid and Low graded wines value counts
```

```python
df_quality = df['quality'].value_counts()
df_quality = df_quality.to_frame()
df_quality.reset_index(inplace=True)
df_quality.rename(columns={'index':'class'}, inplace=True)
df_quality.rename(columns={'quality':'number_of_wines'}, inplace=True)
df_quality.rename(columns={'class':'quality'}, inplace=True)
df_quality.sort_values(by=['quality'], inplace=True)
df_quality
```

```python
plt.figure(figsize=(7, 6), dpi=100) # width and height in inches
plot = sns.barplot(y=df_quality.iloc[:,0], x=df_quality.iloc[:,1], orient='h')  
plot.set_xlabel("Number of wines",fontsize=25)
plot.set_ylabel("Quality grade",fontsize=25)
plot.tick_params(labelsize=17)
plt.savefig('foo.png', transparent=True, bbox_inches='tight')
plt.show() 
```

```python
wine_marks = {'Low':[df_quality.iloc[0,1]+df_quality.iloc[1,1]],'Medium':[df_quality.iloc[2,1]+df_quality.iloc[3,1]],'High':[df_quality.iloc[4,1]+df_quality.iloc[5,1]]}
wine_marks = {'Quality':['Low','Medium','High'], 'Amount':[df_quality.iloc[0,1]+df_quality.iloc[1,1], df_quality.iloc[2,1]+df_quality.iloc[3,1], df_quality.iloc[4,1]+df_quality.iloc[5,1]]}
wine_marks
# df_quality[]
```

```python
df_quality_generalized = pd.DataFrame(data=wine_marks)
df_quality_generalized.iloc[:,0]
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi = 50)

patches, texts, pcts = ax.pie(df_quality_generalized.iloc[:,1], labels=df_quality_generalized.iloc[:,0], autopct='%.1f%%',
    wedgeprops={'linewidth': 2.5, 'edgecolor': 'white'},
    textprops={'size': 'x-large'},
    startangle=-12)
# For each wedge, set the corresponding text label color to the wedge's
# face color.
for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())
plt.setp(pcts, color='white')
plt.setp(texts, fontweight=600)
ax.set_title('Wine Quality Distribution', fontsize=20)
plt.tight_layout()
plt.savefig('foo.png', transparent=True, bbox_inches='tight')
```

```python
plt.figure(figsize=(18, 6), dpi=100)
sns.kdeplot(
   data=df, x="sulphates", hue="quality",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,
)
```

```python
fig, axes = plt.subplots(3, 4, figsize=(16,8))
fig.suptitle('Wine anaylisis barplots')

i = 0
for col in df.columns:
    sns.kdeplot(ax = axes[i//4,i%4],
       data=df, x=col, hue="quality",
       fill=True, common_norm=False, palette="crest",
       alpha=.5, linewidth=0, warn_singular=False
    )
    i = i + 1
    
fig.tight_layout()
```

```python
plt.figure(figsize=(8, 5), dpi=100)
plot = sns.kdeplot(
   data=df, x="alcohol", hue="quality",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,
)
plot.set_xlabel("Alcohol",fontsize=18)
plot.set_ylabel("Density distribution ",fontsize=18)
plot.tick_params(labelsize=13)
plt.savefig('foo.png', transparent=True, bbox_inches='tight')
plt.show() 
```

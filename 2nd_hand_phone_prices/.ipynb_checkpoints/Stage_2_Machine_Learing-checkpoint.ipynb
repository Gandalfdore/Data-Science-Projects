## Polynomial Regression

Certain features of the function seem to have non-line relationship with the target variable

For example the Battery size seems to have an ~ x^2 relationship with the target variable

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#specify degree of 2 for polynomial regression model
#include bias=False means don't force y-intercept to equal zero
poly = PolynomialFeatures(degree=2, include_bias=False)

#reshape data to work properly with sklearn
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

#fit polynomial regression model
poly_reg_model = LinearRegression()
poly_reg_model.fit(X_train_poly, y_train)

#display model coefficients
print(poly_reg_model.intercept_, poly_reg_model.coef_,'\n')

print('Score:',poly_reg_model.score(X_test_poly, y_test))
```

```python
X_train_poly.shape
```

```python
yhat_train = poly_reg_model.predict(X_train_poly)
yhat_test = poly_reg_model.predict(X_test_poly)
```

```python
poly_reg_model.score(X_train_poly, y_train)
```

```python
poly_reg_model.score(X_test_poly, y_test)
```

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 6
    height = 5
    plt.figure(figsize=(width, height))

    ax1 = sns.kdeplot(RedFunction, color="r", label=RedName)
    ax2 = sns.kdeplot(BlueFunction, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    plt.show()
    plt.close()
```

```python
DistributionPlot(y_test, yhat_test, "Actual Values (Test)", "Predicted Values (Test)", "Density distribution of Actual vs Predicted values")
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge

model = Ridge()

model.fit(X_train, y_train)

print(f'model score on training data: {model.score(X_train, y_train)}')
print(f'model score on testing data: {model.score(X_test, y_test)}')
```

```python
coefs = pd.DataFrame(
   model.coef_,
   columns=['Coefficients'], index=X_df.columns
)

coefs.plot(kind='barh', figsize=(9, 7))
plt.title('Ridge model')
plt.axvline(x=0, color='.5')
plt.subplots_adjust(left=.3)

# the magnitudes of the bars under show us the dependencies between a specific feature and the target when all other features remain constant, i.e., conditional dependencies.
```

```python
yhat_test = model.predict(X_test)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(X_train, y_train)

print(f'model score on training data: {model.score(X_train, y_train)}')
print(f'model score on testing data: {model.score(X_test, y_test)}')
```

```python
importances = model.feature_importances_
```

```python
import numpy as np
```

```python
indices = np.argsort(importances)

fig, ax = plt.subplots()
ax.barh(range(len(importances)), importances[indices])
ax.set_yticks(range(len(importances)))
_ = ax.set_yticklabels(np.array(X_cleaned_df.columns)[indices])
```
